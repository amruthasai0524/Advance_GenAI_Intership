# Input Validation Example

```python


valid_audiences = [
    "beginner",
    "intermediate",
    "expert"
]

valid_tones = [
    "formal",
    "casual",
    "fun"
]

def validate_inputs(audience, tone):

    if audience not in valid_audiences:
        audience = "beginner"

    if tone not in valid_tones:
        tone = "formal"

    return audience, tone

audience, tone = validate_inputs(
    "student",
    "friendly"
)

print(audience)
print(tone)
```

# Generate Prompt Function

```python
def generate_prompt(
    topic,
    audience,
    tone,
    style
):

    audience, tone = validate_inputs(
        audience,
        tone
    )

    if style not in styles:
        return "Invalid style"

    prompt = styles[style]

    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

result = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

print(result)
```

# Final Pipeline Example

```python
def final_pipeline(
    topic,
    audience,
    tone,
    style
):

    final_prompt = generate_prompt(
        topic,
        audience,
        tone,
        style
    )

    return final_prompt

output = final_pipeline(
    topic="Generative AI",
    audience="beginner",
    tone="casual",
    style="teaching"
)

print(output)
```
