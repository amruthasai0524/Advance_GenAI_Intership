## Task 4 – ChatPromptTemplate System

Objective

Build a role-based chat prompting system.

```python
from langchain_core.prompts import (
    PromptTemplate,
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate
)
```

```python
system_template = SystemMessagePromptTemplate.from_template(
    "You are a helpful {role}."
)

human_template = HumanMessagePromptTemplate.from_template(
    "Teach me about {topic}"
)

chat_prompt = ChatPromptTemplate.from_messages(
    [system_template, human_template]
)
```

# example

```python
messages = chat_prompt.format_messages(
    role="teacher",
    topic="Neural Networks"
)

print(messages)
```

# another example

```python
messages2 = chat_prompt.format_messages(
    role="motivator",
    topic="Career Growth"
)

print(messages2)
```

# Task 5 – Input Validation Layer

```python
valid_audiences = ["beginner", "intermediate", "expert"]
valid_tones = ["formal", "casual", "fun"]
```

# validation Function

```python
def validate_inputs(audience, tone):

    if audience not in valid_audiences:
        print("Invalid audience detected.")
        audience = "beginner"

    if tone not in valid_tones:
        print("Invalid tone detected.")
        tone = "formal"

    return audience, tone
```

# Testing Validation

```python
audience, tone = validate_inputs("student", "exciting")

print(audience)
print(tone)
```

# Task 6 – Prompt Generator App

# Create Templates Dictionary

```python
styles = {

    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style"
    ),

    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Create interview questions on {topic} for {audience} in a {tone} style"
    ),

    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style"
    )
}
```

# Prompt Generator Function

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style not in styles:
        return "Invalid style selected"

    prompt = styles[style]

    return prompt.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
```

# Test Example

```python
result = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)

print(result)
```
