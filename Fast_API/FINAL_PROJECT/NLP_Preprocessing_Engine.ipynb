# Build Advanced Preprocessing Function

```python
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):

    # Error handling
    if not isinstance(text, str):
        return [], ""

    if text.strip() == "":
        return [], ""

    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove emojis and special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words
    exceptions = ["no", "not"]

    cleaned_tokens = []

    for word in tokens:

        if len(word) > 2 or word in exceptions:

            # Lemmatization
            word = lemmatizer.lemmatize(word)

            cleaned_tokens.append(word)

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

# Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this",

    # Extra test cases
    "",
    "😂😂😂😂😂",
    "123456789",
    "Sooooo coooool brooooo",
    "EMAIL me at abc@gmail.com"
]
```

# Display Results

```python
all_tokens = []

for i, sentence in enumerate(test_sentences, 1):

    tokens, cleaned = preprocess_text(sentence)

    all_tokens.extend(tokens)

    print("=" * 60)
    print(f"Sentence {i}")
    print("Original Text:")
    print(sentence)

    print("\nCleaned Tokens:")
    print(tokens)

    print("\nCleaned Sentence:")
    print(cleaned)
```

# Frequency Analysis

```python
word_freq = Counter(all_tokens)

print("\nTop 10 Most Frequent Words\n")

for word, count in word_freq.most_common(10):
    print(word, ":", count)

print("\nTop 5 Least Frequent Words\n")

least_common = word_freq.most_common()[:-6:-1]

for word, count in least_common:
    print(word, ":", count)
```

# Build Full Pipeline

```python
def full_pipeline(text_list):

    final_tokens = []

    clean_sentences = []

    for text in text_list:

        tokens, cleaned = preprocess_text(text)

        final_tokens.extend(tokens)

        clean_sentences.append(cleaned)

    return {
        "tokens": final_tokens,
        "clean_sentences": clean_sentences
    }
```

# Test Full Pipeline

```python
result = full_pipeline(test_sentences)

print(result)
```
