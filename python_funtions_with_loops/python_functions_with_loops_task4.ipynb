# Problem 1: Mobile Recharge Validation System
# =========================================================

```python
print("Problem 1: Mobile Recharge Validation System\n")

valid_plans = [199, 299, 399, 599]

# Function to validate recharge
def validate_recharge(amount, plan):
    if amount >= 50 and plan in valid_plans:
        return True
    return False

# Retry until valid recharge
while True:
    amount = int(input("Enter recharge amount: "))
    plan = int(input("Enter recharge plan: "))

    if validate_recharge(amount, plan):
        print("Recharge Successful")
        break
    else:
        print("Invalid recharge. Please try again.")
```

# Problem 2: Inventory Reorder Alert System
# =========================================================


```python
print("\n--------------------------------------------")
print("Problem 2: Inventory Reorder Alert System\n")

inventory = {
    "Rice": 10,
    "Oil": 20,
    "Sugar": 8,
    "Soap": 18
}

# Function to check stock
def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(f"{product} → Reorder Alert")
        else:
            print(f"{product} → Stock OK")

check_inventory(inventory)
```

# =========================================================
# Problem 3: Student Result Processing System
# =========================================================

```python
print("\n--------------------------------------------")
print("Problem 3: Student Result Processing System\n")

marks = [65, 70, 45, 80, 60]

# Function to calculate average and result
def calculate_result(student_marks):

    total = 0

    for mark in student_marks:
        total += mark

    average = total / len(student_marks)

    print(f"Average Marks: {average}")

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

result = calculate_result(marks)

print("Result:", result)
```

# Problem 4: Cab Fare Estimator with Retry Option
# =========================================================

```python
print("\n--------------------------------------------")
print("Problem 4: Cab Fare Estimator\n")

# Function to calculate cab fare
def calculate_fare(distance, peak_hour):

    base_fare = 50
    fare = base_fare + (distance * 12)

    # Add 25% extra during peak hour
    if peak_hour.lower() == "yes":
        fare += fare * 0.25

    return fare

# Retry option using while loop
while True:

    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)

    print(f"Total Fare: ₹{total_fare:.2f}")

    retry = input("Do you want to calculate again? (yes/no): ")

    if retry.lower() != "yes":
        break
```

# Problem 5: Employee Attendance Eligibility Checker
# =========================================================


```python
print("\n--------------------------------------------")
print("Problem 5: Employee Attendance Eligibility Checker\n")

attendance = ["P", "A", "P", "P", "A", "P", "P"]

# Function to check attendance eligibility
def check_eligibility(attendance_list):

    present_count = 0

    for status in attendance_list:
        if status == "P":
            present_count += 1

    percentage = (present_count / len(attendance_list)) * 100

    print(f"Attendance Percentage: {percentage:.2f}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

eligibility = check_eligibility(attendance)

print("Eligibility Status:", eligibility)
```

# Problem 6: Password Strength Checker
# =========================================================

```python
print("\n--------------------------------------------")
print("Problem 6: Password Strength Checker\n")

# Function to check password strength
def check_password(password):

    special_characters = "@#$"

    has_digit = False
    has_special = False

    for char in password:

        if char.isdigit():
            has_digit = True

        if char in special_characters:
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

password = input("Enter Password: ")

print(check_password(password))


```
