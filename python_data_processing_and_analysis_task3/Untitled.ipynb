Problem Statement 1: Employee Performance Bonus Eligibility
=========================================================

```python
print("Problem 1: Employee Performance Bonus Eligibility\n")

employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

# Find highest performance score

highest_score = max(employees.values())
# Find employees with highest score

top_performers = []

for employee, score in employees.items():
    if score == highest_score:
        top_performers.append(employee)
# Display result

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

Problem Statement 2: Search Query Keyword Analysis
=========================================================

```python
print("\n--------------------------------------------")
print("Problem 2: Search Query Keyword Analysis\n")

query = "Buy mobile phone buy phone online"

# Convert query to lowercase
query = query.lower()

# Import string module for punctuation removal
import string

# Remove punctuation
for punctuation in string.punctuation:
    query = query.replace(punctuation, "")

# Split sentence into words
words = query.split()

# Create dictionary for frequency count
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Store words occurring more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

# Display result
print(result)
```

# Problem Statement 3: Sensor Data Validation
# =========================================================


```python
print("\n--------------------------------------------")
print("Problem 3: Sensor Data Validation\n")

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Store even readings with index
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

# Display valid readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```
