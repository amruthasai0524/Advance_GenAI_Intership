```python
# Problem 1: Social Media – Post Engagement Analyzer

def analyze_engagement(likes):
    total_likes = sum(likes)

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


likes_list = [200, 150, 300, 400]

analyze_engagement(likes_list)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker

```python


def rainfall_checker(rainfall_data, required_level):

    average_rainfall = sum(rainfall_data) / len(rainfall_data)

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average_rainfall)
    print("Rainfall Status:", status)


rainfall = [70, 75, 68, 80, 67]

rainfall_checker(rainfall, 70)
```

# Problem 4: Social Media – Duplicate Account Detection

```python


def duplicate_accounts(usernames):

    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


accounts = ["amrutha", "reddy", "amrutha", "data_science"]

duplicate_accounts(accounts)
```

# Problem 6: Agriculture – Premium Crop Price Filter

```python


def premium_crop_filter(prices):

    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


crop_prices = [1500, 2500, 1800, 3200, 1000]

premium_crop_filter(crop_prices)
```

# Problem 9: E-Learning – Student Attendance Eligibility System

```python


def attendance_eligibility(attendance):

    total_classes = len(attendance)
    attended = 0

    for status in attendance:
        if status == "P":
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


attendance_data = ["P", "P", "A", "P", "P"]

attendance_eligibility(attendance_data)
```

# Problem 10: Smart Electricity Bill Analyzer

```python


def electricity_bill(units):

    if units <= 100:
        bill = units * 3

    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"

    elif bill <= 1500:
        usage = "Moderate Usage"

    else:
        usage = "High Usage"

    print("Electricity Bill:", bill)
    print("Usage Status:", usage)


electricity_bill(250)
```
