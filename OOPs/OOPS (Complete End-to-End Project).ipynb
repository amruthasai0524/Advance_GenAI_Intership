```python
# =========================================================
# Problem 1: Smart Light Controller
# =========================================================

class SmartLight:
    
    # Constructor
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"
    
    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"
    
    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"
    
    # Method to display status
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Creating object
light = SmartLight("Bedroom Light")

# Turning ON the light
light.turn_on()

# Displaying status
light.display_status()
```

# =========================================================
# Problem 2: Employee ID Card System
# =========================================================


```python
class Employee:
    
    # Constructor
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    # Method to display employee details
    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Creating object
employee1 = Employee("Rahul", "EMP102", "IT")

# Displaying employee ID card
employee1.display_id_card()
```

# =========================================================
# Problem 3: Mobile Contact Record
# =========================================================


```python
class Contact:
    
    # Constructor
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    # Method to display contact information
    def save_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# Creating object
contact1 = Contact("Anita", "9876543210")

# Displaying contact details
contact1.save_contact()
```

# Problem 4: Product Price Tag Generator
# =========================================================

```python
class Product:
    
    # Constructor
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price
    
    # Method to display price tag
    def display_price_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")


# Creating object
product1 = Product("Headphones", 2499)

# Displaying product price tag
product1.display_price_tag()
```

# Problem 5: Movie Rating Display System
# =========================================================

```python
class Movie:
    
    # Constructor
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating
    
    # Method to display movie details
    def display_movie_details(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating} / 5")


# Creating object
movie1 = Movie("Inception", 4.8)

# Displaying movie details
movie1.display_movie_details()
```

# Problem 6: Delivery Address Manager
# =========================================================

```python
class Delivery:
    
    # Constructor
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    # Method to display delivery details
    def display_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# Creating object
delivery1 = Delivery("Suman", "Hyderabad")

# Displaying delivery details
delivery1.display_delivery_details()
```
