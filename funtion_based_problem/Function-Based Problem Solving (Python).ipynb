# Smart Parking Lot Management System

```python


def parking_management(capacity, logs):
    parked = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            parked -= 1

        # Track peak usage
        if parked > peak_usage:
            peak_usage = parked

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak_usage)

    if parked > capacity:
        print("Parking Status: Full")
        print("Alert: Parking capacity exceeded!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_management(capacity, vehicle_logs)
```

# Online Food Delivery Time Estimator

```python


def estimate_delivery_time(distance, traffic, weather):

    # Base time calculation
    eta = distance * 5

    # Traffic conditions
    if traffic == "High":
        eta += 10
    elif traffic == "Medium":
        eta += 5

    # Weather conditions
    if weather == "Rainy":
        eta += 5
    elif weather == "Storm":
        eta += 15

    print("Estimated Delivery Time:", eta, "minutes")


# Sample Input
distance = 8
traffic_level = "High"
weather_condition = "Rainy"

estimate_delivery_time(distance, traffic_level, weather_condition)
```

# Movie Theatre Seat Occupancy Analyzer

```python


def seat_occupancy(total_seats, booked_seats):

    booked_count = len(booked_seats)

    occupancy_percentage = (booked_count / total_seats) * 100

    print("Occupancy:", occupancy_percentage, "%")

    # Show status
    if occupancy_percentage == 100:
        print("Show Status: Housefull")
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Open Additional Shows")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200

# 150 booked seats
booked_seats = [1] * 150

seat_occupancy(total_seats, booked_seats)
```

# Cloud Server Load Classification System

```python


def classify_server_load(cpu_readings):

    average_cpu = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", average_cpu, "%")

    # Classification
    if average_cpu < 50:
        print("Server Status: Normal")
    elif 50 <= average_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

# Smart Classroom Resource Usage Monitor

```python


def monitor_resources(resource_usage):

    overused_resources = []

    # Threshold for overuse
    threshold = 8

    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused_resources.append(resource)

    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(resource_usage)
```

```python
# Online Event Registration Capacity Controller

def event_registration_controller(capacity, registrations):

    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_capacity = 100
registrations = 105

event_registration_controller(event_capacity, registrations)
```
